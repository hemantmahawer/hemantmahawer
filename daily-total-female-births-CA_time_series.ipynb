# Decomposing

```
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
df = pd.read_csv(r'/content/drive/MyDrive/daily-total-female-births-CA.csv', header=0, parse_dates=[0])
```

```
df.head()
```

```
df.index = df["date"]
```

```
result = seasonal_decompose(df['births'], model='additive')
```

```
result.plot()
```

```
result2 = seasonal_decompose(df['births'], model='multiplicative')
```

```
result2.plot()
```

# differencing

```
df = pd.read_csv(r'/content/drive/MyDrive/daily-total-female-births-CA.csv', header=0, parse_dates=[0])
```

```
df.head()
```

```
df['lag1']= df['births'].shift(1)
```

```
df['births_diff_1']=df['births'].diff(periods=1)
```

```
df.head()
```

```
df.index = df["date"]
result_a = seasonal_decompose(df['births'], model='additive')
result_a.plot()
```

```
df.index = df["date"]
result_b = seasonal_decompose(df.iloc[1:,3], model='additive')
result_b.plot()
```

```
df['births'].plot()
```

```
df['births_diff_1'].plot()
```

```
df['births_diff_2'] = df['births_diff_1'].diff(periods =12)
```

```
df['births_diff_2'].plot()
```

```
result_c = seasonal_decompose(df.iloc[7:,3], model='additive')
result_c.plot()
```
